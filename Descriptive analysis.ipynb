```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

import pickle
```

```python
df_purchase=pd.read_csv('purchase data.csv')
```

```python
df_purchase.head()
```

```python
df_purchase.isnull().sum()
```

```python
scaler=pickle.load(open('scaler.pickle','rb'))
```

```python
pca=pickle.load(open('pca.pickle','rb'))
```

```python
kmeans_pca=pickle.load(open('kmeans_pca.pickle','rb'))
```

```python
features=df_purchase[df_purchase.columns.values[-7:]]
```

```python
df_purchase_segm_std=scaler.transform(features)
```

```python
df_purchase_segm_pca=pca.transform(df_purchase_segm_std)
```

```python
purchase_segm_kmeans_pca=kmeans_pca.predict(df_purchase_segm_pca)
```

```python
df_purchase_predictor=df_purchase.copy()
```

```python
df_purchase_predictor['Segment']=purchase_segm_kmeans_pca
```

```python
df_purchase_predictor
```

## Data analysis by Segments

Data analysis by customer

```python
df_purchase_predictor.head()
```

```python
temp1= df_purchase_predictor[['ID','Incidence']].groupby(['ID'],as_index=False).count()
temp1= temp1.set_index('ID')
temp1= temp1.rename(columns={'Incidence':'N_visits'})
```

```python
temp1.head()
```

```python
temp2= df_purchase_predictor[['ID','Incidence']].groupby(['ID'],as_index=False).sum()
temp2= temp2.set_index('ID')
temp2= temp2.rename(columns={'Incidence':'N_Purchases'})
```

```python
temp3=temp1.join(temp2)
```

```python
temp3.head()
```

```python
temp3['Average_N_purchases']=temp3['N_Purchases']/temp3['N_visits']
```

```python
temp4= df_purchase_predictor[['ID','Segment']].groupby(['ID'],as_index=False).mean()
temp4= temp4.set_index('ID')
df_purchase_descr=temp3.join(temp4)
```

```python
df_purchase_descr.head()
```

```python
segm_prop=df_purchase_descr[['N_Purchases','Segment']].groupby('Segment').count()/df_purchase_descr.shape[0]
segm_prop=segm_prop.rename(columns={'N_Purchases':'Segment_Proportion'})
segm_prop.head()
```

```python
plt.figure(figsize=(10,8))
plt.pie(segm_prop['Segment_Proportion'],
       labels=['Standard','Career-Focused','Fewer-Opportunity','Wealthy'],
       autopct='%1.3f%%',
       colors=['g','b','r','y'])
```

```python
segment_mean=df_purchase_descr.groupby(['Segment']).mean()
```

```python
segment_mean
```

```python
segment_std=df_purchase_descr.groupby(['Segment']).std()
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=[0,1,2,3],
       tick_label=['Standard','Career-Focused','Fewer-Opportunity','Wealthy'],
       height=segment_mean['N_visits'],
       yerr=segment_std['N_visits'],
       color=['g','b','r','y'])
plt.xlabel('Segment')
plt.ylabel('Number of store visits')
plt.title('Number of store visits by segment')
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=[0,1,2,3],
       tick_label=['Standard','Career-Focused','Fewer-Opportunity','Wealthy'],
       height=segment_mean['N_Purchases'],
       yerr=segment_std['N_Purchases'],
       color=['g','b','r','y'])
plt.xlabel('Segment')
plt.ylabel('Number of store purchases')
plt.title('Number of store purchases by segment')
```

```python
segment_mean.head()
```

```python
plt.figure(figsize=(10,8))
plt.bar(x=[0,1,2,3],
       tick_label=['Standard','Career-Focused','Fewer-Opportunity','Wealthy'],
       height=segment_mean['Average_N_purchases'],
       yerr=segment_std['Average_N_purchases'],
       color=['g','b','r','y'])
plt.xlabel('Segment')
plt.ylabel('Average number of purchases')
plt.title('Avg n purchases by segment')
```

## Brand choice

```python
temp.head()
```

```python
temp.drop('bruh',axis=1)
```

```python
segment_brand_revenue[['Segment','Revenue Brand 1']]=temp[['Segment','Revenue Brand 1']].groupby('Segment',as_index=False).sum()
```

```python
segment_brand_revenue
```
